Zero pad all spectrogram with an offset of 2

```python
for i, spectrogram in enumerate(spectrograms):
    spectrograms[i] = np.pad(spectrogram, ((2,2),(0,0)), "constant", constant_values=0)
```

Select all frequencies at timestep 0 with a context size of 2 in recording 0

```python
offset = 2
context = 2

### Get index pair from index map (1-2 lines)
i, j = (0, 2)

### Calculate starting timestep using offset and context (1 line)
start_j = j + offset - context

## Calculate ending timestep using offset and context (1 line)
end_j = j + offset + context + 1

### Get data at index pair with context (1 line)
frames = spectrograms[i][start_j:end_j,:]

### Return data at index pair with context and label at index pair (1 line)
frames
```
