Import the numpy module.

```python
import numpy as np
```

Load array of spectrograms to environment.

```python
spectrograms  = np.load("./samples/spectrograms/linear/all_spectrograms.npy", allow_pickle=True)
```

Select recording 0.

```python
spectrograms[0]
```

Select recording 1.

```python
spectrograms[1]
```

Select all frequencies at timestep 2 in recording 0.

```python
spectrograms[0][2, :]
```

Select all frequencies from timestep 1 to timestep 3 (exclusive) in recording 1.

```python
spectrograms[1][1:3, :]
```

Iterate and print the indecies of all recordings

```python
for i, _ in enumerate(spectrograms):
    print(i)
```

Iterate and print the shape of all recordings.

```python
for i, spectrogram in enumerate(spectrograms):
    print(i, spectrogram.shape)
```

Iterate and print the indecies of all timesteps and shape of each frame in recording 0.

```python
for i, frame in enumerate(spectrograms[0]):
    print(i, frame.shape)
```

Iterate and print the indecies of all timesteps in all recordings.

```python
for i, spectrogram in enumerate(spectrograms):
    for j, frame in enumerate(spectrogram):
        print(i, j)
```

Use two for loops to create a list of all (recording, timestep) index pairs.

```python
spectrogram_index_map = []
for i, spectrogram in enumerate(spectrograms):
    for j, frame in enumerate(spectrogram):
        spectrogram_index_pair = (i, j)
        spectrogram_index_map.append(spectrogram_index_pair)
```

Use a for loop to iterate and print the list of all (recording, timestep) index pairs.

```python
for i, spectrogram_index_pair in enumerate(spectrogram_index_map):
    print(i, spectrogram_index_pair)
```
