```python
context = 2
offset = 2

# Define reference (center) index j
j = 0

# Define starting timestep j
start_j = j + offset - context

# Define ending timestep j
end_j = j + offset + context + 1

### Get data at index pair with context (1 line)
frames = spectrogram[start_j:end_j,:]
```
