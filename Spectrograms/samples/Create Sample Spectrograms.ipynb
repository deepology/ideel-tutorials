```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from scipy.io import wavfile
```

---

```python
sample_rate, samples = wavfile.read('./recordings/wav/recording-1.wav')
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate)
np.save("./spectrograms/linear/spectrogram-1.npy", spectrogram.T)
```

```python
plt.pcolormesh(times, frequencies, spectrogram, shading='auto')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
log_spectrogram = np.log(spectrogram+0.0001)
np.save("./spectrograms/log/log_spectrogram-1.npy", log_spectrogram.T)
```

```python
plt.pcolormesh(times, frequencies, log_spectrogram, shading='auto')
plt.ylabel('Log Frequency [Log Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
